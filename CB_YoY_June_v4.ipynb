```python
# dependencies
import pandas as pd
import os
import numpy as np
```

```python
from datetime import datetime
from datetime import date
```

```python
# import year-over-year tripdata
citi_2013 = "data/trip_details/201306-citibike-tripdata.csv"
citi_2014 = "data/trip_details/201406-citibike-tripdata.csv"
citi_2015 = "data/trip_details/201506-citibike-tripdata.csv"
citi_2016 = "data/trip_details/201606-citibike-tripdata.csv"
citi_2017 = "data/trip_details/201706-citibike-tripdata.csv"
citi_2018 = "data/trip_details/201806-citibike-tripdata.csv"
```

```python
# read the CSVs in pandas/convert into dataframes
to_datetime = lambda d: datetime.strptime(d, '%m/%d/%Y %H:%M')
to_datetime2 = lambda d: datetime.strptime(d, '%Y-%m-%d %H:%M:%S')
to_datetime3 = lambda d: datetime.strptime(d, '%m/%d/%Y %H:%M:%S')

citi13_df = pd.read_csv('data/trip_details/201306-citibike-tripdata.csv', converters={'starttime': to_datetime2, 'stoptime': to_datetime2})
print("Done reading 2013 data.")
citi14_df = pd.read_csv('data/trip_details/201406-citibike-tripdata.csv', converters={'starttime': to_datetime2, 'stoptime': to_datetime2})
print("Done reading 2014 data.")
citi15_df = pd.read_csv('data/trip_details/201506-citibike-tripdata.csv', converters={'starttime': to_datetime, 'stoptime': to_datetime})
print("Done reading and converting 2015 data.")
citi16_df = pd.read_csv('data/trip_details/201606-citibike-tripdata.csv', converters={'starttime': to_datetime3, 'stoptime': to_datetime3})
print("Done reading and converting 2016 data.")
citi17_df = pd.read_csv('data/trip_details/201706-citibike-tripdata.csv', converters={'starttime': to_datetime2, 'stoptime': to_datetime2})
print("Done reading 2017 data.")
#citi18_df = pd.read_csv('data/trip_details/201806-citibike-tripdata.csv', converters={'starttime': to_datetime2, 'stoptime': to_datetime2})
#print("Done reading 2018 data.")

print(citi13_df.head(5))
print("-"*20)
print(citi14_df.head(5))
print("-"*20)
print(citi15_df.head(5))
print("-"*20)
print(citi16_df.head(5))
print("-"*20)
print(citi17_df.head(5))


df_total = pd.concat([citi13_df, citi14_df, citi15_df, citi16_df, citi17_df])
print(df_total.head(5))
print("And tail...")
print(df_total.tail(5))
df_total.to_csv("total.csv", sep=',')
```

```python
# read the remaining CSVs in pandas/convert into dataframes
citi18_df = pd.read_csv(citi_2018)
```

```python
citibike_June = [df_total, citi18_df]

citibike_df = pd.concat(citibike_June)
citibike_df.head()
```

```python
# rename columns so there's never spaces in the column name
citibike_df = citibike_df.rename(columns={"tripduration": "Trip_Duration", 
                                          "starttime": "Start_Time", 
                                          "stoptime": "Stop_Time",
                                          "start station id": "Start_Station_ID", 
                                          "start station name": "Start_Station_Name", 
                                          "start station latitude": "Start_Lat.", 
                                          "start station longitude": "Start_Long.",
                                          "end station id": "End_Station_ID", 
                                          "end station name": "End_Station_Name", 
                                          "end station latitude": "End_Lat.", 
                                          "end station longitude": "End_Long.",
                                          "bikeid": "Bike_ID", 
                                          "usertype": "Rider_Type",
                                          "birth year": "Born", 
                                          "gender": "Gender", 
                                         })



citibike_df.head()
```

```python
# number of rows in df
len(citibike_df)
```

```python
#ensure all lines match
citibike_df.count()
```

```python
# clean data by dropping null rows
citibike_clean_df = citibike_df.dropna(how='any')

# verify counts for number of recorded trips
citibike_clean_df.count()
```

```python
# Traverse through gender column and rewrite the value from a number to a word where condition matches. 
citibike_clean_df.Gender[citibike_clean_df.Gender == 1] = "Male"
citibike_clean_df.Gender[citibike_clean_df.Gender == 2] = "Female"
citibike_clean_df.Gender[citibike_clean_df.Gender == 0] = "Unidentified"
print(citibike_clean_df.Gender) 
```

```python
citibike_clean_df.dtypes
```

```python
# in case there are still missing data
citibike_clean_df = citibike_clean_df.fillna(0)
```

```python
citibike_clean_df["End_Station_ID"] = citibike_clean_df["End_Station_ID"].astype(int)
```

```python
citibike_clean_df["Born"] = citibike_clean_df["Born"].apply(lambda x: int(x) if x != '\\N' else 0)
```
